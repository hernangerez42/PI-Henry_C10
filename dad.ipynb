```python
import pandas as pd
df=pd.read_csv("dataset/movies_dataset_trabajado.csv")
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Seleccionar las primeras 10000 filas del dataset
sample_size = 10000
sample = df.head(sample_size)

# Seleccionar las columnas relevantes y crear nuevos índices
selected_columns = ["title", "genres", "overview", "tagline"]
sample = sample[selected_columns]

# Eliminar filas con valores nulos en las columnas seleccionadas
sample = sample.dropna()

# Combinar las columnas relevantes en una sola columna de texto
sample['combined_text'] = sample['title'] + ' ' + sample['genres'] + ' ' + sample['overview'] + ' ' + sample['tagline']

# Preprocesamiento de los textos en la muestra
vectorizer = CountVectorizer()
text_matrix = vectorizer.fit_transform(sample['combined_text'].values.astype('U'))

# Calcular la similitud coseno entre los textos en la muestra
similarity_matrix = cosine_similarity(text_matrix)

def predecir_peliculas_similares(pelicula, n=1):
    # Encontrar el índice de la película dada en la muestra
    matching_movies = sample[sample['title'] == pelicula]
    
    if len(matching_movies) == 0:
        print("No se encontró la película en la muestra.")
        return []

    idx = matching_movies.index[0]

    if idx >= len(similarity_matrix):
        print("El índice de la película está fuera de los límites.")
        return []

    # Obtener los puntajes de similitud coseno para la película dada en la muestra
    similarity_scores = list(enumerate(similarity_matrix[idx]))

    # Ordenar las películas por puntajes de similitud en orden descendente
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

    # Obtener los índices de las películas similares en la muestra (excluyendo la película dada)
    similar_movie_indices = [i for i, _ in similarity_scores if i != idx]

    # Obtener los nombres de las películas similares en la muestra
    similar_movies = sample.loc[similar_movie_indices]['title'][:n].tolist()

    return similar_movies
```

```python
predecir_peliculas_similares("grumpier old men")
```
